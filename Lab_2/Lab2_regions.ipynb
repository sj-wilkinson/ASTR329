```python
def make_plot(i):
    m1 = mean_ss1[i]
    m2 = mean_ss2[i]
    m3 = mean_ss3[i]
    m4 = mean_ss4[i]
    m5 = mean_ss5[i]
    m6 = mean_ss6[i]
    m7 = mean_ss7[i]
    m8= mean_ss8[i]
    m9 = mean_ss9[i]
    m10 = mean_ss10[i]
    s1 = std_ss1[i]
    s2 = std_ss2[i]
    s3 = std_ss3[i]
    s4 = std_ss4[i]
    s5 = std_ss5[i]
    s6 = std_ss6[i]
    s7 = std_ss7[i]
    s8= std_ss8[i]
    s9 = std_ss9[i]
    s10 = std_ss10[i]
    
    means = [m1,m2,m3,m4,m5,m6,m7,m8,m9,m10]
    sds = [s1,s2,s3,s4,s5,s6,s7,s8,s9,s10]
    
    x = [5,10,15,20,25,30,40,50,100]
    
    fig,ax1 = plt.subplots()
    ax1.set_xticks(x)
    ax1.set_xticklabels(('5','10','15','20','25','30','40','50','100x100'),fontsize = 10)
    ax2 = ax1.twinx()
    ax2.set_ylabel('Standard Deviation')
    ax1.set_ylabel('Mean')
    ax2.plot(x,sds[:-1],label = 'Standard Deviation')
    ax1.plot(x,means[:-1], color = 'red', label = 'Mean')
    ax1.legend()
    ax2.legend(loc = 'upper center')
    plt.savefig('subregion_compare_small'+str(i)+'.png')
    plt.show()
```

```python
for i in range(10):
    make_plot(i)
```
